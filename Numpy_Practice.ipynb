```python
import numpy as np

list = [4,3,6,9,0]
print(list)

a = np.array([3,6,8,9,0])
b = np.array([[2,4,6,7,2],[0,7,4,2,1]])
c = np.array([[2,4,6,9,1],[0,1,4,6,4],[6,3,5,8,0]])
print("1d Array",a)
print("2d Array",b)
print("3d Array",c)

print("NO of elements in each array")
print(a.size)
print(b.size)
print(c.size)

print("Type of Data in each Array")
print(a.dtype)
print(b.dtype)
print(c.dtype)

print("No of Rows and Columns")
print(a.shape)
print(b.shape)
print(c.shape)

print("Tranpose of a array")
print(b.transpose())


print("Random Array")
print(np.empty((2,2),dtype =int))

print("Array with 3 rows and 5 columns with 1s")
print(np.ones((3,5),dtype=int))

print("Array with 2 rows and 3 cols with 0's")
print(np.zeros((2,3),dtype =int))

print("Numbers 1 to 10")
print(np.arange(1,10))

print("Numbers 1 to 10 with step 2-odd numbers")
print(np.arange(1,10,2))

print("Numbers 2 to 10 with step 2-even numbers")
print(np.arange(2,10,2))

print(a.reshape((1,5)))

Arr=np.arange(1,51)
print(Arr)
Arr = Arr.reshape(10,5)
print(Arr)
print(Arr[0])
print(Arr[0,0])
print(Arr[2:5])
print(Arr[:,2]) #all rows of 2 column

A = np.arange(1,11).reshape(5,2)
B = np.arange(5,15).reshape(5,2)
print(A)
print(B)
print(A+B)
print(A-B)
print(A/B)
B=B.reshape(2,5)
print(A@B)
print(A.dot(B))
print(B.max())
print(A.min())
print(B.argmax()) #index of max val
print(np.sum(B))
print("Mean of A",np.mean(A))
print("Square of each element of B",np.sqrt(B))
print("Deviation of A",np.std(A))
print("Log of each element of B",np.log(B))

print(np.random.random(1))
print(np.random.random(2))
print(np.random.random((2,2)))
print(np.random.randint(1,10)) #random number from 1 to 10
print(np.random.randint(1,10,(2,2)))
print(np.random.rand(2,2))
print(np.random.randn(2,2))

Opps = np.arange(1,10)
print("Numbers in Opps",Opps)
print(np.random.choice(Opps))

str1 = "Khadeja is my name"
str2= " I am a CS major"
print(np.char.add(str1,str2))
print(np.char.lower(str2))
print(np.char.upper(str1))
print(np.char.split(str1))
str3 = "Ayesha\n what the hell of this"
print(np.char.splitlines(str3))
print(np.char.replace(str2,"major","minor"))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
def rmse(targets, predictions):
    return np.sqrt(np.mean((predictions - targets) ** 2))
```

```python
# Before dropping NaNs
print(weather[weather['Type'].isna()])
```

```python
weather = weather.dropna(subset=['Type', 'Temperature (C)'])
```

```python
# Create inputs and targets
input, target = weather[['Type']], weather['Temperature (C)']

# Create and train the model
model = LinearRegression().fit(input, target)

# Generate predictions
prediction_6 = model.predict(input)

# Compute loss to evalute the model
loss = rmse(target, prediction_6)
print('Loss:', loss)
```

```python
# Create inputs and targets
input_cols = ['Type', 'Humidity','Apparent Temperature (C)','Wind Bearing (degrees)','Visibility (km)','Wind Speed (km/h)','Pressure (millibars)', 'Breezy', 'Breezy and Dry', 'Breezy and Foggy',
        'Breezy and Mostly Cloudy', 'Breezy and Overcast',
        'Breezy and Partly Cloudy', 'Clear',
        'Dangerously Windy and Partly Cloudy', 'Drizzle', 'Dry',
        'Dry and Mostly Cloudy', 'Dry and Partly Cloudy', 'Foggy',
        'Humid and Mostly Cloudy', 'Humid and Overcast',
        'Humid and Partly Cloudy', 'Light Rain', 'Mostly Cloudy',
        'Overcast', 'Partly Cloudy', 'Rain', 'Windy', 'Windy and Dry',
        'Windy and Foggy', 'Windy and Mostly Cloudy', 'Windy and Overcast',
        'Windy and Partly Cloudy' ]
inputs, targets = weather[input_cols], weather['Temperature (C)']

# Create and train the model
model = LinearRegression().fit(inputs, targets)

# Generate predictions
predictions = model.predict(inputs)

# Compute loss to evalute the model
loss = rmse(targets, predictions)
print('Loss:', loss)
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train, inputs_test, targets_train, targets_test = train_test_split(inputs, targets, test_size=0.1)
```

```python
# Create and train the model
model = LinearRegression().fit(inputs_train, targets_train)

# Generate predictions
predictions_test = model.predict(inputs_test)

# Compute loss to evalute the model
loss = rmse(targets_test, predictions_test)
print('Test Loss:', loss)
```

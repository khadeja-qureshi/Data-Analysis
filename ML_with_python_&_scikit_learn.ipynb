Linear Regression

```python
non_smoker = medical[medical.smoker =='no']
```

```python
sns.scatterplot(data=non_smoker,x='age',y='charges',alpha=0.7)
plt.title("Age vs Charges")
```

The values in the "age" column of the dataset are called the inputs to the model and the values in the charges column are called "targets"

charges=wÃ—age+b  w=weight and b=bias

```python
def estimate_charges(age, w, b):
    return w * age + b
```

```python
w = 50
b = 100
```

```python
estimate_charges(40,w,b)
```

```python
ages = non_smoker.age
ages
```

```python
estimated_charges = estimate_charges(ages,w,b)
estimated_charges
```

```python
target = non_smoker.charges
target
```

```python
plt.plot(ages, estimated_charges,'r-o');
plt.xlabel('Age');
plt.ylabel('Estimated Charges');
```

```python
def try_parameters(w, b):
    ages = non_smoker.age
    target = non_smoker.charges

    estimated_charges = estimate_charges(ages, w, b)

    plt.plot(ages, estimated_charges, alpha=0.9)
    plt.scatter(ages, target,alpha=0.8)
    plt.xlabel('Age');
    plt.ylabel('Charges')
    plt.legend(['Estimate', 'Actual'])
```

```python
try_parameters(200,600)
```

```python
try_parameters(400,4000)
```

```python
import pandas as pd

list = [6,5,4,0,2]
A = pd.Series(list)
print(A)
print(pd.Series(list,index=["a","b","c","d","e"],name="Numbers"))

Scalar_Series = pd.Series(0.5)
print(Scalar_Series)
print(pd.Series(0.5,index = [1,2,3]))

dict_series = {1:'r',2:'s',3:'v'}
Series = pd.Series(dict_series)
print(Series)
print(Series[1])
print(Series[0:2])
print(max(Series))
dict_series={'r':[1,2,3],'s':[5,6,7]}
Series = pd.Series(dict_series)
print(Series)


L1 = [1,6,8,9]
print(pd.DataFrame(L1))
L2 = [[2,4,5,6],[9,0,6,7]]
print(pd.DataFrame(L2))

Random = [{'a':1,"b":2,"c":3,"e":4,"f":5}]
print(pd.DataFrame(Random))

CSV = pd.read_csv("gld.csv")
print(CSV)
print(CSV.head()) #5 rows
print(CSV.columns)
print(CSV.shape)
print(CSV.size) #cells
print(CSV.head(2))
print(CSV.tail(2))
print(CSV.describe())
print(CSV.info())

Marks = pd.read_csv("sample.csv")
print(Marks)
print(Marks.head())
print(Marks.isnull())
print(Marks.isnull().sum())
print(Marks.isnull().sum().sum())

print(Marks.shape)
Marks2 = Marks.dropna() #drops rows
print(Marks2.shape)
Marks2 = Marks.dropna(axis =1) #drops cols
print(Marks2.shape)
Marks.dropna(how ='any') #any row with 1 null val
Marks.dropna(how = "all") #any row with all null val

print(Marks.fillna(0))
print(Marks.fillna({'Physics': 'none'}))
Physics_mean = Marks['Physics'].mean()
print(Marks.fillna({'Physics': Physics_mean}))

print(Marks.replace(to_replace=33.0, value=30.0))
print(Marks['Maths'].replace(to_replace=[50.0,51.0],value=['A-','A+']))

Subject = pd.read_csv("Sample2.csv",index_col=['Roll No.'])
print(Subject.head())
print(Subject.loc[1])
print(Subject.loc[[5,6,7]])
print(Subject.loc[5:15,'Chemistry'])
print(Subject.loc[Subject['Physics']<50])
print(Subject.loc[Subject['Physics']>80,['Maths']])

print(Subject.iloc[0])
print(Subject.iloc[[0]])
print(Subject.iloc[0:5,1:4])

Branch_groups = Subject.groupby(by = "Branch")
print(Branch_groups.groups)

Two_Branch = Subject.groupby(by = ["Branch",'Chemistry'])
print(Two_Branch.groups)

Score1 = pd.DataFrame({'RollNo':[1,2,3,4],'Chemistry':[89.9,54,34.9,86.5]})
Score2 = pd.DataFrame({'RollNo':[1,2,3,4],'Physics':[34.6,43.8,78.0,54]})
print(pd.merge(Score1,Score2))

Score3 = pd.DataFrame({'RollNo':[3,4,7,8],'Maths':[23,98.0,56.8,34]})
print(pd.merge(Score1,Score3))
print(pd.merge(Score1,Score3,how="left"))
print(pd.merge(Score2,Score3,how="right"))
print(pd.merge(Score2,Score3,how="outer"))
print(pd.concat([Score1,Score2]))

```
